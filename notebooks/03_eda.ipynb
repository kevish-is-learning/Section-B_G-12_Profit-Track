# 03 Exploratory Data Analysis (EDA)
This notebook explores the cleaned dataset to identify key patterns, trends, and relationships to answer: **How can management increase sales, profit margin, and stock efficiency?**

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')

# Configure visualizations
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette('muted')

df = pd.read_csv('data/processed/cleaned_dataset.csv')
df['order_date'] = pd.to_datetime(df['order_date'])
df['ship_date'] = pd.to_datetime(df['ship_date'])
df.head()
```

## 1. Univariate Analysis
Understanding the distribution of key numerical and categorical variables.

```
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
sns.histplot(df['sales'], bins=50, kde=True, ax=axes[0], color='blue')
axes[0].set_title('Sales Distribution')
axes[0].set_xlim(0, df['sales'].quantile(0.95)) # Cap for visibility

sns.histplot(df['profit'], bins=50, kde=True, ax=axes[1], color='green')
axes[1].set_title('Profit Distribution')
axes[1].set_xlim(df['profit'].quantile(0.05), df['profit'].quantile(0.95))

sns.countplot(data=df, x='product_category', ax=axes[2], palette='Set2')
axes[2].set_title('Orders by Product Category')
plt.tight_layout()
```

## 2. Bivariate Analysis
Analyzing the relationships between variables, focusing on drivers of Sales and Profit.

```
fig, axes = plt.subplots(1, 2, figsize=(16, 5))
sns.barplot(data=df, x='region', y='sales', estimator=np.sum, errorbar=None, ax=axes[0], palette='viridis')
axes[0].set_title('Total Sales by Region')

sns.barplot(data=df, x='region', y='profit', estimator=np.sum, errorbar=None, ax=axes[1], palette='magma')
axes[1].set_title('Total Profit by Region')
plt.tight_layout()
```

```
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='product_category', y='profit', hue='customer_segment', showfliers=False)
plt.title('Profit Distribution across Categories by Customer Segment')
plt.show()
```

## 3. Time Series Analysis
Examining seasonality and revenue trends.

```
monthly_sales = df.set_index('order_date').resample('ME')['sales'].sum().reset_index()

plt.figure(figsize=(14, 6))
sns.lineplot(data=monthly_sales, x='order_date', y='sales', marker='o', color='crimson')
plt.title('Monthly Total Sales Trend')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.show()
```

## Key Insights from EDA
1. **Regional Driver**: Certain regions dominate overall sales but might not have proportional profitability.
2. **Category Performance**: Profitability varies heavily across categories like Furniture vs. Technology.
3. **Seasonality**: Sales peak during specific times of the year, providing a clear opportunity to minimize inventory stockout.
